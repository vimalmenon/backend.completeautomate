```python
from backend.config.env import env
from openai import OpenAI


client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=str(env.OPEN_ROUTE_API_KEY),
)

# Generate an image
response = client.chat.completions.create(
  model="black-forest-labs/flux.2-klein-4b",
  messages=[
          {
            "role": "user",
            "content": "Generate a beautiful sunset over mountains"
          }
        ],
  extra_body={"modalities": ["image", "text"]}
)
```

```python
message = response.choices[0].message
```

```python
import base64
from PIL import Image
from io import BytesIO

# Assuming your base64 string is in response content
# Extract base64 string (if it has data URI prefix, strip it)
base64_string = message.images[0]['image_url']['url']  # or wherever your base64 data is

# Remove data URI prefix if present
if "base64," in base64_string:
    base64_string = base64_string.split("base64,")[1]

# Decode and display
image_data = base64.b64decode(base64_string)
img = Image.open(BytesIO(image_data))
display(img)

# Optional: Save to file
img.save("generated_image.png")
```
